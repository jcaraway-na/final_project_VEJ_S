```python
# Import dependencies
import pandas as pd
import numpy as np
```

```python
# Import csv
v0_df = pd.read_csv('../../Vision_Zero_Crash_Report_Data.csv')
```

```python
# List columns
v0_df.columns
```

```python
# View nulls
v0_df.isna().sum()
```

```python
# Drop unnecessary columns
v0_df = v0_df.drop(columns=['case_id', 'rpt_latitude', 'rpt_longitude', 'rpt_block_num', 'rpt_street_pfx', 'rpt_street_name', 'rpt_street_sfx', 'street_name', 'street_nbr', 'street_name_2', 'street_nbr_2', 'contrib_factr_p1_id', 'contrib_factr_p2_id','units_involved','atd_mode_category_metadata','apd_confirmed_fatality', 'onsys_fl', 'private_dr_fl'])
```

```python
# Drop rows without latitude & longitude
v0_df = v0_df.dropna(subset=["point"])
v0_df.isna().sum()
```

```python
# Drop time from date time column
v0_df['crash_date'] = v0_df['crash_date'].str.split(' ',expand=True)[0]
```

```python
# Turn crash date to datetime
v0_df['crash_date'] = pd.to_datetime(v0_df['crash_date'])
```

```python
# Encode Y/N columns
y_n = ['crash_fatal_fl', 'road_constr_zone_fl']
v0_df[y_n] = v0_df[y_n].replace({'Y': 1, 'N': 0})
```

```python
# Fill blank values in "flag" columns with 0, switch 'Y' to 1
flag_columns = ['pedestrian_fl', 'motor_vehicle_fl', 'motorcycle_fl', 'bicycle_fl', 'other_fl']
v0_df[flag_columns] = v0_df[flag_columns].replace({np.nan: '0', 'Y': 1})
v0_df.head()
```

```python
# Drop NAs
v0_df = v0_df.dropna()
```

```python
# Total deaths by type
death_columns = ['motor_vehicle_death_count', 'bicycle_death_count', 'pedestrian_death_count', 'motorcycle_death_count', 'other_death_count']
death_list = []
for column in death_columns:
    death_list.append(v0_df[column].sum())
    print(f"{column}: {v0_df[column].sum()}")
```

```python
total_crashes = v0_df['crash_id'].count()
```

```python
total_fatal = v0_df['crash_fatal_fl'].sum()
percent_fatal = (total_fatal / total_crashes) * 100
death_list.append(total_fatal)
death_list.append(percent_fatal)
death_list
```

```python
# Total injuries by type
injury_columns = ['motor_vehicle_serious_injury_count', 'bicycle_serious_injury_count', 'pedestrian_serious_injury_count', 'motorcycle_serious_injury_count', 'other_serious_injury_count']
injury_list = []
for column in injury_columns:
    injury_list.append(v0_df[column].sum())
    print(f"{column}: {v0_df[column].sum()}")
```

```python
total_injury = v0_df['sus_serious_injry_cnt'].sum()
percent_injury = (total_injury / total_crashes) * 100
injury_list.append(total_injury)
injury_list.append(percent_injury)
```

```python
data = {'deaths': death_list, 'injuries': injury_list}
index = ['motor_vehicle', 'bicycle', 'pedestrian', 'motorcycle', 'other', 'total', 'percent']
totals_df = pd.DataFrame(data=data, index=index)
totals_df
```

```python
# Separate data into pre- and post-March 2020
precovid_df = v0_df.copy()
precovid_df = precovid_df[precovid_df['crash_date'] < '2020-03-15']
precovid_df
```

```python
postcovid_df = v0_df.copy()
postcovid_df = postcovid_df[postcovid_df['crash_date'] > '2020-03-15']
postcovid_df
```

```python
# Export pre-covid data to csv
precovid_df.to_csv('precovid_data.csv', index=False)
```
